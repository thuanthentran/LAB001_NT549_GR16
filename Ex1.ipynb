## Part 2. Object-Oriented Programming (OOP) with Python

````python
# Example of a simple class to represent a network router
# This is a basic implementation to demonstrate object-oriented programming (OOP) concepts in Python.

class Router:
    def __init__(self, name, ip_address):
        self.name = name
        self.ip_address = ip_address
        self.connected_devices = []

    def connect_device(self, device_name):
        self.connected_devices.append(device_name)
        print(f"{device_name} has been connected to {self.name}.")

    def disconnect_device(self, device_name):
        if device_name in self.connected_devices:
            self.connected_devices.remove(device_name)
            print(f"{device_name} has been disconnected from {self.name}.")
        else:
            print(f"{device_name} is not connected to {self.name}.")

    def show_connected_devices(self):
        print(f"Devices connected to {self.name}:")
        for device in self.connected_devices:
            print(f"- {device}")
# Example usage
router1 = Router("Home Router", "192.168.1.1")
router1.connect_device("Laptop")
router1.connect_device("Smartphone")
router1.show_connected_devices()
router1.disconnect_device("Laptop")
router1.show_connected_devices()
````

# Exercise 1: Maximize

You are given a function:

$$
f(X) = X^2
$$

You are also given $K$ lists. The $i^{th}$ list consists of $N_i$ elements.

You have to pick **one element from each list** such that the value from the equation below is maximized:

$$
S = \left(f(X_1) + f(X_2) + \dots + f(X_K)\right) \bmod M
$$

Where:

- $X_i$ denotes the element picked from the $i^{th}$ list.
- $\bmod$ denotes the **modulo operator**.

Find the maximum possible value $S_{max}$.

> Note:  
> You must pick **exactly one element from each list**, not necessarily the largest element.  
> The squares of the selected elements are summed, and then the modulo operation is applied.

---

# Input Format

- The first line contains two space-separated integers:

$$
K \quad M
$$

- The next $K$ lines each contain:
  - An integer $N_i$ denoting the number of elements in the $i^{th}$ list.
  - Followed by $N_i$ space-separated integers representing the elements of that list.

---

# Constraints

$$
1 \le K \le 7
$$

$$
1 \le M \le 1000
$$

$$
1 \le N_i \le 7
$$

$$
1 \le |a_{i,j}| \le 10^9
$$

Where $a_{i,j}$ is the $j^{th}$ element in the $i^{th}$ list.

---

# Output Format

Output a single integer denoting the value:

$$
S_{max}
$$

---

# Sample Input

```
3 1000
2 5 4
3 7 8 9
5 5 7 8 9 10
```

---

# Sample Output

```
206
```

---

# Explanation

Choosing:

- $5$ from the **1st list**
- $9$ from the **2nd list**
- $10$ from the **3rd list**

gives:

$$
S = (5^2 + 9^2 + 10^2) \bmod 1000
$$

$$
= (25 + 81 + 100) \bmod 1000
$$

$$
= 206
$$

This is the **maximum possible value**.

## Method and ideas:
1. Parse all the arguments which are usable for the solution: M, and N[i] of each list below
2. Use 'product' from 'itertools' for automatically calculating the S = (f(X_1) + f(X_2) + ... + f(X_k))%M where f(X_i) = X_i^2 for each combination of the lists
3. Return the max combination

````python
from itertools import product

K, M = map(int, input().split())

lists = []

for _ in range (K):
    arr = list(map(int, input().split()))
    lists.append(arr[1:])

max_value = 0

for combination in product(*lists):
    total = sum(x*x for x in combination)
    value = total%M
    if value > max_value:
        max_value = value
        
print (max_value)
````

# Method of implementation, algorithm and approach to solve the problem:
1. Use 'datetime' library for parsing time and time zone format.
2. Use abs and 'total_seconds()' to calculate and show the difference between the two days in second format.

````python
#!/bin/python3

import math
import os
import random
import re
import sys
from datetime import datetime

# Complete the time_delta function below.
def time_delta(t1, t2):
    time_format = "%a %d %b %Y %H:%M:%S %z"
    
    dt1 = datetime.strptime(t1, time_format)
    dt2 = datetime.strptime(t2, time_format)
    
    diff = abs(int((dt1 - dt2).total_seconds()))
    
    return str(diff)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        t1 = input()

        t2 = input()

        delta = time_delta(t1, t2)

        fptr.write(delta + '\n')

    fptr.close()
````

# Exercise 3: Write a function

An extra day is added to the calendar almost every four years as February 29, and the day is called a leap day. It corrects the calendar for the fact that our planet takes approximately 365.25 days to orbit the sun. A leap year contains a leap day.

In the Gregorian calendar, three conditions are used to identify leap years:

- The year can be evenly divided by 4, is a leap year, unless:
    - The year can be evenly divided by 100, it is NOT a leap year, unless:
        - The year is also evenly divisible by 400. Then it is a leap year.
This means that in the Gregorian calendar, the years 2000 and 2400 are leap years, while 1800, 1900, 2100, 2200, 2300 and 2500 are NOT leap years. Source

**Task**

Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean True, otherwise return False.

Note that the code stub provided reads from STDIN and passes arguments to the is_leap function. It is only necessary to complete the is_leap function.

**Input Format**

Read, the $year$ to test.

**Constraints**
$$
1900 \le year \le 10^5
$$
**Output Format**

The function must return a Boolean value (True/False). Output is handled by the provided code stub.

**Sample Input 0**
$$
1990
$$
**Sample Output 0**
$$
False
$$

**Explanation 0**

1990 is not a multiple of 4 hence it's not a leap year.


# Method of implementation, algorithm and approach to solve the problem.
Use sample conditional blocks and basic operators

````python
def is_leap(year):
    leap = False
    
    if (year % 4 == 0 and year % 100 != 0):
        leap = True
    elif (year % 400 == 0):
        leap = True
    else:
        leap = False
    return leap
````

### **Exercise 4**: Capitalized!

You are asked to ensure that the first and last names of people begin with a capital letter in their passports. For example, `alison heck` should be capitalised correctly as `Alison Heck`.

`alison heck` $\Rightarrow$ `Alison Heck`

Given a full name, your task is to capitalize the name appropriately.

**Input Format**

A single line of input containing the full name, $S$.

**Constraints**

* $0 < len(S) < 1000$
* The string consists of alphanumeric characters and spaces.

**Note:** in a word only the first character is capitalized. Example 12abc when capitalized remains 12abc.

**Output Format**

Print the capitalized string, $S$.

**Sample Input**

```text
chris alan
```

**Sample Output**

```text
Chris Alan

### Method of implementation, algorithm and approach to solve the problem.

Iterate through each word in the input string, capitalize the first character of each word, then join the words back into a complete string.

````python
def solve(s):
    words = s.split(" ")
    result = []

    for word in words:
        if word:
            result.append(word[0].upper() + word[1:])
        else:
            result.append(word)

    return " ".join(result)
````

### **Exercise 5**: Score Words

In this challenge, the task is to debug the existing code to successfully execute all provided test files.

Consider that vowels in the alphabet are `a`, `e`, `i`, `o`, `u` and `y`.

Function `score_words` takes a list of lowercase words as an argument and returns a score as follows:

The score of a single word is 2 if the word contains an even number of vowels. Otherwise, the score of this word is 1. The score for the whole list of words is the sum of scores of all words in the list.

Debug the given function `score_words` such that it returns a correct score.

Your function will be tested on several cases by the locked template code.

**Input Format**

The input is read by the provided locked code template. In the first line, there is a single integer $n$ denoting the number of words. In the second line, there are $n$ space-separated lowercase words.

**Constraints**

* $1 \le n \le 20$
* Each word has at most 20 letters and all letters are English lowercase letters

**Output Format**

The output is produced by the provided and locked code template. It calls function `score_words` with the list of words read from the input as the argument and prints the returned score to the output.

**Sample Input 0**

```text
2
hacker book
```

**Sample Output 0**

4

**Explanation 0**

There are two words in the input: hacker and book. The score of the word hacker is $2$ because it contains an even number of vowels, i.e. $2$ vowels, and the score of book is $2$ for the same reason. Thus the total score is $2 + 2 = 4$.

**Sample Input 1**

```text
3
programming is awesome
```

**Sample Output 1**

4

**Explanation 1**

There are $3$ words in the input: programming, is and awesome. The score of programming is $1$ since it contains $3$ vowels, an odd number of vowels. The score of is is also $1$ because it has an odd number of vowels. The score of awesome is $2$ since it contains $4$ vowels, an even number of vowels. Thus, the total score is $1 + 1 + 2 = 4$.

### Method of implementation, algorithm and approach to solve the problem.

For each word in the list, count the number of vowels. If the number of vowels is even, add 2 to the score; otherwise add 1. Sum the scores of all words.

````python
def is_vowel(letter):
    return letter in ['a', 'e', 'i', 'o', 'u', 'y']

def score_words(words):
    score = 0
    for word in words:
        num_vowels = 0
        for letter in word:
            if is_vowel(letter):
                num_vowels += 1
        if num_vowels % 2 == 0:
            score += 2
        else:
            score += 1
    return score
````

#### **Exercise 6**: No Idea!

There is an array of *n* integers. There are also 2 **disjoint sets**, *A* and *B*, each containing *m* integers. You like all the integers in set *A* and dislike all the integers in set *B*. Your initial happiness is 0. For each *i* integer in the array, if $i \in A$, you add 1 to your happiness. If $i \in B$, you add -1 to your happiness. Otherwise, your happiness does not change. Output your final happiness at the end.

**Note:** Since *A* and *B* are sets, they have no repeated elements. However, the array might contain duplicate elements.

**Constraints**

$1 \le n \le 10^5$

$1 \le m \le 10^5$

$1 \le \text{Any integer in the input} \le 10^9$

**Input Format**

The first line contains integers *n* and *m* separated by a space.
The second line contains *n* integers, the elements of the array.
The third and fourth lines contain *m* integers, *A* and *B*, respectively.

**Output Format**

Output a single integer, your total happiness.

**Sample Input**

```text
3 2
1 5 3
3 1
5 7
```

**Sample Output**

```text
1

### Method of implementation, algorithm and approach to solve the problem.

Iterate through each element in the array. If the element is in set A, increase happiness by 1; if it is in set B, decrease happiness by 1. After processing all elements, output the final happiness value.

````python
n, m = map(int, input().split())

arr = list(map(int, input().split()))
A = set(map(int, input().split()))
B = set(map(int, input().split()))

happiness = 0

for num in arr:
    if num in A:
        happiness += 1
    elif num in B:
        happiness -= 1

print(happiness)
````
